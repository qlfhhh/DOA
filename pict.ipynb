```python
import numpy as np
import matplotlib.pyplot as plt
```

```python

plt.rcParams['font.family'] = 'SimHei' # 正常显示中文
plt.rcParams['axes.unicode_minus'] = False # 正常显示负号
```

```python
x = [-5,-4,-3,-2,-1]
sr = [-24.3,-25.02,-28,-32,-34.5]
r=[-15.1,-16.7,-18,-20.4,-21.8]
se = [-21.5,-24.1,-26.3,-26.9,-27.7]
e=[-15.8,-18,-19.9,-21.1,-23.9]
cnn=[-10.9,-12.7,-15.6,-16.5,-20.2]
plt.plot(x,sr,label="本方法+R-MUSIC")
plt.plot(x,r,label="R-MUSIC")
plt.plot(x,se,label="本方法+ESPRIT")
plt.plot(x,e,label="ESPRIT")
plt.plot(x,cnn,label="CNN")
plt.xticks(x)
plt.yticks([-30,-25,-20,-15])
plt.xlabel("SNR")
plt.ylabel("MPSE[dB]")
plt.legend(loc="lower left")
plt.show()
```
